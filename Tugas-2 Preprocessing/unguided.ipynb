2. Cek nilai duplikat

```python
# Memeriksa nilai duplikat  
int(df.duplicated().sum())
```

3. Menemukan null values buat presentase

```python
# Analisis nilai hilang dalam DataFrame
null_analysis = df.isnull().sum().to_frame('Null Values')
null_analysis['Total Values'] = df.count()
null_analysis['Percentage Null (%)'] = (null_analysis['Null Values'] / len(df)) * 100

# Menampilkan kolom dengan nilai hilang
print(null_analysis[null_analysis['Null Values'] > 0])
```

4. Drop missing value berdasarkan baris, kolom, imputasi dengan mean, modus, median

```python
#mencari berapa banyak nilai yang hilang
imputate = df.select_dtypes(include=[np.number])
imputate.isna().sum()
```

```python
# Drop rows with missing values
data_drop_row = df.dropna()
```

```python
# Drop columns with missing values
data_drop_columns = df.dropna(axis=1)
```

```python
# Impute with mean
data_impute_mean = df.copy()
data_impute_mean['Time_taken'] = data_impute_mean['Time_taken'].fillna(data_impute_mean['Time_taken'].mean())

data_impute_mean.isna().sum()
```

```python
# Impute with median
data_impute_median = df.copy()
data_impute_median['Time_taken'] = data_impute_median['Time_taken'].fillna(data_impute_median['Time_taken'].median())

data_impute_median.isna().sum()
```

```python
# Impute with modus
data_impute_modus = df.copy()
data_impute_modus['Time_taken'] = data_impute_modus['Time_taken'].fillna(data_impute_modus['Time_taken'].mode()[0])

data_impute_modus.isnull().sum()
```
